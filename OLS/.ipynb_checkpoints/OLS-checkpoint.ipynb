# OLS Implementation from scratch

### OLS

```python
class OLS():

  def __init__(self):
    self.b = 0
    self.m = 0

  def fit(self, X, y):
    n = X.shape[0]
    self.m = (n*np.sum(X*y) - np.sum(y)*np.sum(X))/(n*np.sum(X**2) - np.sum(X)**2)
    self.b = y.mean() - X.mean() * self.m

  def predict(self, X):
    return np.dot(X,self.m) + self.b
```

```python
ols = OLS()
ols.fit(X,y)
print(f'm:{ols.m}, b:{ols.b}')
```

```python
X_ = np.linspace(X.min() - 1.5, X.max() + 1.5, 100)
y_pred = ols.predict(X_)

fig, ax = plt.subplots()
ax.scatter(X,y, color = 'limegreen', marker= 'o')
ax.plot(X_, y_pred, color = 'orangered', label = f'$y = {ols.m:.3f}x + {ols.b:.3f}$')
ax.set_xlim(0, 12)
ax.set_ylim(0, y_pred.max())
ax.set_xlabel(X.name)
ax.set_ylabel(y.name)
ax.legend()
ax.set_title('Salary vs Experience')
plt.show()
fig.savefig('ols_scatter.png')
```

```python
class OLSVec():

    def __init__(self):
        pass

    def fit(self, X, y):
        X = np.column_stack((np.ones(len(X)), X))
        self.b = np.linalg.inv(X.T @ X) @ X.T @ y
    
    def predict(self, X):
        return np.dot(X, self.b[1:]) + self.b[0]
```

```python
olsvec = OLSVec()
olsvec.fit(X,y)
print(f'b:{olsvec.b}')
```

```python
X_ = np.linspace(X.min() - 1.5, X.max() + 1.5, 100)
X_ = X_.reshape(-1, 1)
y_pred = olsvec.predict(X_)

fig, ax = plt.subplots()
ax.scatter(X,y, color = 'limegreen', marker= 'o')
ax.plot(X_, y_pred, color = 'orangered', label = f'$y = {olsvec.b[1]:.3f}x + {olsvec.b[0]:.3f}$')
ax.set_xlim(0, 12)
ax.set_ylim(0, y_pred.max())
ax.set_xlabel(X.name)
ax.set_ylabel(y.name)
ax.legend()
ax.set_title('Salary vs Experience')
plt.show()
fig.savefig('ols_scattervec.png')
```

## Evaluating our model

```python
def mse(y_true,y_pred):
    n = len(y_true)
    error = y_true - y_pred
    return sum(error ** 2) / n
```

```python
def rmse(y_true,y_pred):
    n = len(y_true)
    error = y_true - y_pred
    mse =  sum(error ** 2) / n
    return np.sqrt(mse)
```

```python
def mae(y_true,y_pred):
    n = len(y_true)
    error = y_true - y_pred
    return np.sum(np.abs(error)) / n
```

```python
def r2(y_true, y_pred):
    mean_y_true = np.mean(y_true)
    ss_total = np.sum((y_true - mean_y_true)**2)
    ss_residual = np.sum((y_true - y_pred)**2)
    r_squared = 1 - (ss_residual / ss_total)
    return r_squared
```

```python
y_pred_vec = olsvec.predict(X_copy)
y_pred_ols = ols.predict(X.values)
```

```python
print('OLS Metrics:\n'
      f'MSE: {mse(y, y_pred_ols):.3f}\n'
      f'RMSE: {rmse(y, y_pred_ols):.3f}\n'
      f'MAE: {mae(y, y_pred_ols):.3f}\n'
      f'R2: {r2(y, y_pred_ols):.3f}')

print('OLSVEC Metrics:\n'
      f'MSE: {mse(y, y_pred_vec):.3f}\n'
      f'RMSE: {rmse(y, y_pred_vec):.3f}\n'
      f'MAE: {mae(y, y_pred_vec):.3f}\n'
      f'R2: {r2(y, y_pred_vec):.3f}')
```

# Using machine learning libraries

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X_copy,y)
print(f'm:{linear_reg.coef_}, b:{linear_reg.intercept_}')
```

```python
X_ = np.linspace(X.min() - 1.5, X.max() + 1.5, 100)
y_pred = linear_reg.predict(X_.reshape(-1,1))

fig, ax = plt.subplots()
ax.scatter(X,y, color = 'limegreen', marker= 'o')
ax.plot(X_, y_pred, color = 'orangered', label = f'$y = {linear_reg.coef_[0]:.3f}x + {linear_reg.intercept_:.3f}$')
ax.set_xlim(0, 12)
ax.set_ylim(0, y_pred.max())
ax.set_xlabel(X.name)
ax.set_ylabel(y.name)
ax.legend()
ax.set_title('Salary vs Experience')
plt.show()
fig.savefig('linreg_scatter.png')
```

```python
y_pred_linreg = linear_reg.predict(X_copy)
print('Linear Regression Metrics:\n'
      f'MSE: {mean_squared_error(y, y_pred_linreg):.3f}\n'
      f'RMSE: {np.sqrt(mean_squared_error(y, y_pred_linreg)):.3f}\n'
      f'MAE: {mean_absolute_error(y, y_pred_linreg):.3f}\n'
      f'R2: {r2_score(y, y_pred_linreg):.3f}')
```
