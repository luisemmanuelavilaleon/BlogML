```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
plt.style.use('seaborn-v0_8')
```

```python
h = 1e-6
```

```python
class Gradient_Descent():
    
    def gradient_descent(self, func, n_vars, n_iter = 20, alpha = 0.01):
        x_ni = []
        x_n = np.random.normal(loc = 0, scale = 2, size=n_vars)
        x_ni.append(x_n)
        for i in range(n_iter):
            z_n = func(x_n)
            x_h = [self.sum_h(x_n, j) for j in range(n_vars)]
            grad = np.array([(func(x_h[j]) - func(x_n)) / h for j in range(n_vars)])
            x_n = x_n - alpha * grad
            x_ni.append(x_n)
        return np.array(x_ni)
        
    def sum_h(self, array, i):
        arr = np.copy(array)
        arr[i] += h
        return arr    
```

```python
f = lambda x : x[0]**2 +  x[1]**2 + x[0] * np.sin(x[1]) + x[1] * np.sin(x[0])
f2 = lambda x : (x[0] ** 2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2
```

```python
gd = Gradient_Descent()
min = gd.gradient_descent(f2, 2, n_iter=10, alpha=0.01)
```

```python
x = np.linspace(-4, 4, 400)
y = np.linspace(-4, 4, 400) 
X = np.meshgrid(x, y)
Z = f2(X)
```

```python
fig, ax = plt.subplots()
contour = ax.contourf(X[0],X[1] , Z, levels=20, cmap='viridis')
ax.scatter(min[:,0], min[:,1], color = 'orangered', marker = 'o')
for i in range(min.shape[0] - 1):
    ax.annotate('', xy = (min[i + 1, 0], min[i + 1,1]), xytext = (min[i, 0], min[i, 1]), arrowprops = dict(width = 1, headwidth = 4, facecolor = 'black'))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Curvas de Nivel de f(x, y)')
ax.grid(False)
fig.colorbar(contour, ax=ax, label='f(x, y)')
plt.show()
```
